# scikit-plot

*scikit-plot* is an extension of [matplotlib](http://matplotlib.org/) for datascientist. Proposed graphs are a frequent need when playing with data.

[documentation](https://scikit-plot.readthedocs.io/en/stable/) [source](https://github.com/reiinakano/scikit-plot) [installation](https://scikit-plot.readthedocs.io/en/stable/Quickstart.html?highlight=installation) [tutorial](https://scikit-plot.readthedocs.io/en/stable/Quickstart.html?highlight=installation#your-first-plot)

```python
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
data= load_iris()
X, y = data.data, data.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
import scikitplot as skplt
from sklearn.linear_model import LogisticRegression
nb = LogisticRegression(solver='lbfgs', multi_class='ovr')
nb = nb.fit(X_train, y_train)
y_probas = nb.predict_proba(X_test)
skplt.metrics.plot_roc(y_test, y_probas);
```
