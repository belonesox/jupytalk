## example

```python
%matplotlib inline
```

```python
from ggplot import ggplot, diamonds, aes, geom_point, scale_color_brewer, xlab, ylab
from ggplot import ggtitle

try:
    ggplot(diamonds, aes(x='carat', y='price', color='cut')) + \
        geom_point() +\
        scale_color_brewer(type='diverging', palette=4) + \
        xlab("Carats") + ylab("Price") + ggtitle("Diamonds")
except ImportError as e:
    print(e)
```
