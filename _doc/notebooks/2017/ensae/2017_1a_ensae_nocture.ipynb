En somme, ce projet m’a appris certaines choses qui sont valables partout : je pensais que c’était plus important de faire progresser le code que de progresser moi-même. Mais je me suis rendu compte que ça ne servait à rien de se mettre avec le meilleur pour un super résultat si l’on est incapable ensuite de le faire tout seul.*
